```python
import zipfile

# Specify the name of the uploaded zip file
zip_file_name = 'outputs_v2.zip'  # Replace with the name of your zip file

# Extract the contents of the zip file
with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall('/content/outputs_v2_outer')  # Specify the destination folder

print("Zip file extracted successfully.")
```

```python
import os

# Get the list of files in the /content directory
folder_path = '/content/outputs_v2_outer/outputs_v2'
files_dir = os.listdir(folder_path)

# Print the list of files
for file in files_dir:
    print(file)
```

```python
len(files_dir)
```

```python
# Sort the file list
sorted_files = sorted(files_dir)

if len(sorted_files) > 0:
    # Get the path of the first file
    first_file_path = os.path.join(folder_path, sorted_files[0])

    # Open and process the first file
    with open(first_file_path, 'r') as file:
        first_file_content = file.read()

    # Print the content of the first file
    print(first_file_content)
else:
    print("No files found in the folder.")
```

```python
# Embedding

!pip install -q sentence-transformers
!pip install -q langchain

from langchain.embeddings import HuggingFaceEmbeddings
import torch

model_name = "sentence-transformers/all-MiniLM-L6-v2" # all-MiniLM-L6-v2: structure of SBert
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': False}
hf = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
first_file_embedding = torch.tensor(hf.embed_documents(first_file_content))
first_file_embedding
```

```python
first_file_embedding_data = {
        "file_name": sorted_files[0],
        "file_embedding": first_file_embedding.tolist()
    }
```

```python
import json

# Define the file path
first_file_embedding_path = '/content/first_file_embedding_data.json'  # Replace with your desired file path

# Save the dictionary as a JSON file
with open(first_file_embedding_path, 'w') as file:
    json.dump(first_file_embedding_data, file)

print("Dictionary saved as a JSON file.")

# Doanload the embedding data
files.download(first_file_embedding_path)
```
